```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"..\raw\chip-dataset-cleaned.csv",
                 delimiter=",")
df.info()
```

```python
df = df.drop(df.columns[[10,11,12]],axis=1)
df.head()
```

## N Chips Manufactured

```python
df_w = df.copy()
```

```python
chip_count = df_w['Vendor'].value_counts()
chip_count
```

```python

type(chip_count)
```

```python
colors = ['#007F99', '#0099CC', '#33B5E5', '#66CCFF', '#99E6FF']

fig, ax = plt.subplots(figsize=(10, 5))
chart = chip_count.plot(kind='bar', color=colors, edgecolor='black', width=0.6)

chart.bar_label(chart.containers[0], label_type='edge', fontsize=10, color='black', weight='bold')

plt.title('Chips Manufactured (CPU+GPU)', fontfamily='Arial', fontsize=14, fontweight='bold', color='#005f73')
plt.xlabel('')
chart.set_xticklabels(chip_count.index, rotation=0, fontsize=11, fontfamily='Arial', color='#005f73')
plt.ylabel('Number of Chips', fontsize=12, fontfamily='Arial', color='#005f73')
plt.grid(axis='y', linestyle='--', alpha=0.7)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

## GPU Process Size Trend

```python
df_ps = df_w[['Type','Date','Process_Size']].where(df_w['Type']=='GPU').dropna()
df_ps
```

```python
df_ps['Date'] = pd.to_datetime(df_ps['Date'])
df_ps['Year'] = df_ps['Date'].dt.year
df_ps 
```

```python
df_ps = df_ps.groupby('Year')['Process_Size'].mean().reset_index()
df_ps
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(df_ps['Year'], df_ps['Process_Size'], color='#00B5E2', linewidth=2, marker='o', markersize=6, markerfacecolor='#007ACC')

plt.title('GPU Process Size Trend', fontsize=14, fontweight='bold', color='#003f5c')
plt.xlabel('Year', fontsize=12, fontfamily='Arial', color='#003f5c')
plt.ylabel('Process Size (nm)', fontsize=12, fontfamily='Arial', color='#003f5c')
plt.xticks(fontsize=10, fontfamily='Arial', color='#003f5c')
plt.yticks(fontsize=10, fontfamily='Arial', color='#003f5c')

plt.grid(True, linestyle='--', alpha=0.7)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.show()
```
